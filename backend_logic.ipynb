```python
PIZZA= 18.99
CALZONE = 12.99
VEGAN_OPTION = 2.50
CHEESE_BREAD = 5.99
GARLIC_BREAD = 4.99
CINNAMON_STICKS = 3.59
COOKIES = 3.59
POP = 2.00
BOTTLED_WATER = 1.50
MEAL_AND_DRINK_COMBO = 0.1
DINNER_COMBO = 0.2
TAX = 0.13
DELIVERY_FEE = 5.99
DISTANCE_SURCHARGE = 2.00

def discount(subtotal:float,discount_amount:float):
    discount_amt = subtotal * discount_amount
    discounted_total = subtotal - discount_amt
    return discounted_total


def delivery_distance_premium(cx_distance):      
    if int(cx_distance)<=10:
        charge=  DELIVERY_FEE+ 0
    elif 10<int(cx_distance)<=19:
        charge= DELIVERY_FEE+ 2
    elif 20<int(cx_distance)<=29:
        charge=  DELIVERY_FEE+ 4
    else:
        charge= None
       
    return charge
    



def tip(subtotal:float,tip_percentage:float):
    
    tip_amount= subtotal * tip_percentage/100
    return tip_amount



def tax(subtotal:float):
    tax_amount= subtotal* TAX
    return tax_amount



def is_valid_option(Order_Choice:str):
    Valid_Choices = {'meal','appetizer','dessert','beverages','done'}
    return Order_Choice.lower() in Valid_Choices


def customer_order ():
    subtotal=0
    order_meal=False
    order_appetizer=False
    order_dessert=False
    order_beverages=False
    while True:
        input_choice = str(input('What would you like to order? OPTIONS: meal,appetizer,dessert,beverages,done\n'))
        if  is_valid_option(input_choice):
            Order_Choice = input_choice.lower()
            
            if Order_Choice=='done':
                if subtotal==0:
                    print('You don\'t seem to order anything. Please try again later.')
                    return 0
                break

            if Order_Choice=='meal':
                order_meal=True
                meal_choice=str(input('What meal would you like to order? OPTIONS: pizza, calzone:\n'))
                quantity=int(input('How many would you like?\n'))
                vegan=input('Do you want vegan? y/n \n').lower()
                
                if meal_choice=='pizza':
                    if vegan=='y':
                        subtotal += PIZZA * quantity + VEGAN_OPTION
                    else:
                        subtotal += PIZZA * quantity
                        
                         
                elif meal_choice== 'calzone':
                    if vegan=='y':
                       subtotal+= CALZONE * quantity + VEGAN_OPTION
                    else:
                        subtotal += CALZONE * quantity
                

            elif Order_Choice == 'appetizer':
                order_appetizer=True
                appetizer_option= str(input('What bread would you like to add ? OPTIONS: Cheese bread, Garlic bread\n')).lower()
                quantity=int(input('How many bread would you like?\n'))
                if appetizer_option=='cheese bread':
                    subtotal+= CHEESE_BREAD * quantity
                elif appetizer_option=='garlic bread':
                    subtotal += GARLIC_BREAD *quantity
                                 
                    
                    
                
            elif Order_Choice== 'dessert':
                order_dessert=True
                dessert_option= str(input('What do you like for dessert? OPTIONS: Cinnamon sticks, Cookies\n ')).lower() 
                quantity=int(input('How many dessert would you like?\n'))
                if dessert_option=='cinnamon sticks':
                    subtotal+= CINNAMON_STICKS * quantity
                elif dessert_option=='cookies':
                    subtotal+= COOKIES * quantity
            
            elif Order_Choice=='beverages':
                order_beverages=True
                beverage_option= input('What beverages would you like? OPTIONS: Pop,Bottled water\n').lower()
                quantity=int(input('How many beverages would you like?\n'))
                if beverage_option=='pop':
                    subtotal+= POP* quantity
                elif beverage_option=='bottled water':
                    subtotal += BOTTLED_WATER * quantity
        
       



            another=input('Would you like to order another item? y/n \n').lower()
            if another=='n':
                break
        
        else:
            print('Sorry, but that option not recognized\n Please try again!')

    
    
    discount_amt=0        
    if  order_meal and order_appetizer and order_dessert and order_beverages :
        discount_amt+= DINNER_COMBO
        print('You ordered bit of everything! Please enjoy 20% discount')
                
                
    elif order_meal and order_beverages:
        discount_amt += MEAL_AND_DRINK_COMBO
        print('You ordered a meal and drink! Please enjoy your 10% discount')
    
    if discount_amt>0:
        subtotal_with_discount= discount(subtotal,discount_amt)    
        print(f'Your subtotal prior to discount was:{subtotal:.2f}')
        print(f'Your subtotal after the discount was:{subtotal_with_discount:.2f}')
        return subtotal_with_discount
    
    else:
        return subtotal


    


def main():
    print('Thanks for visiting The Pizza Haven! \n Please note: there is a fuel surcharge for longer trips\n')
    cx_distance = int(input('How many kilometers are you from the restaurant?\n'))
    delivery_charge = delivery_distance_premium(cx_distance)
    
    if delivery_charge is None:
        print('Sorry, but we don\'t deliver that far away.')
        return

    print(f'The delivery charge is: {delivery_charge}')
    still_interested = input('Would you still like to place an order with The Pizza Haven? y/n \n').lower()
    if still_interested == 'n':
        print('We understand. Have a good day!')
        return
    
    subtotal_with_discount = customer_order()
    if subtotal_with_discount > 0:
        tip_entered = float(input('Please enter a tip percentage for the driver: \n'))

        # Rounding components before calculating total
        subtotal_rounded = round(subtotal_with_discount, 2)
        tax_amount = round(tax(subtotal_with_discount), 2)
        tip_final = round(tip(subtotal_with_discount, tip_entered), 2)
        delivery_charge_rounded = round(delivery_charge, 2)

        total = round(subtotal_rounded + tax_amount + tip_final + delivery_charge_rounded, 2)

        print(f'SUBTOTAL: {subtotal_rounded:.2f}')
        print(f'TAX: {tax_amount:.2f}')
        print(f'TIP: {tip_final:.2f}')
        print(f'DELIVERY: {delivery_charge_rounded:.2f}')
        print(f'TOTAL: {total:.2f}')

# Run the program
main()
```
